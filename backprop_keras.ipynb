```python
import numpy as np
import pandas as pd

```

```python
df = pd.DataFrame([[6,8,9],[7,9,9],[6,10,6],[5,12,7]], columns=['cgpa','profile_score','lpa'])
df
```

```python
import tensorflow
from tensorflow import keras  
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
#This means: ğŸ‘‰ The layer has 2 neurons
model.add(Dense(2, input_dim=2, activation='linear'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_weights = [np.array([[0.1, 0.1], [0.1, 0.1]], dtype=np.float32), np.array([0, 0], dtype=np.float32), np.array([[0.1], [0.1]], dtype=np.float32), np.array([0], dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizers = keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=optimizers, loss='mean_squared_error')
```

```python
df['lpa'].values
```

```python
model.fit(df.iloc[:, 0:-1], df['lpa'].values, epochs=75, verbose=1,batch_size=1)
```
